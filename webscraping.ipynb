```python
import undetected_chromedriver as uc
import time
import random
import csv
from selenium.webdriver.common.by import By

# 1. إعداد المتصفح "الشبح" مع تحديد الإصدار
options = uc.ChromeOptions()
options.add_argument("--window-size=1366,768")
options.add_argument("--disable-gpu")

# تحديد الإصدار 146 ليتوافق مع جهازك
try:
    driver = uc.Chrome(options=options, version_main=146)
except Exception as e:
    print(f"Error starting Chrome: {e}")
    # إذا فشل، نحاول تشغيله بدون تحديد إصدار كخيار احتياطي
    driver = uc.Chrome(options=options)

# 2. ملف CSV - تقسيم متطور للعناوين
csv_file = open('yellowpages_final_separated.csv', 'w', newline='', encoding='utf-8-sig')
writer = csv.writer(csv_file)
# التقسيم الجديد: اسم، محافظة، منطقة، عنوان، تليفون
writer.writerow(['Restaurant_Name', 'Governorate', 'District', 'Address_Details', 'Phone'])

def process_address(address_text):
    """تقسيم العنوان بناءً على الفواصل المعتادة في يلوبيدجز"""
    parts = [p.strip() for p in address_text.split(',')]
    
    # الافتراض: العنوان غالباً يكون (تفاصيل، منطقة، محافظة)
    gov = parts[-1] if len(parts) > 0 else "N/A"
    dist = parts[-2] if len(parts) > 1 else "N/A"
    # باقي الأجزاء هي تفاصيل العنوان
    addr_details = ", ".join(parts[:-2]) if len(parts) > 2 else address_text
    
    return gov, dist, addr_details

def scrape_data():
    try:
        cards = driver.find_elements(By.CLASS_NAME, "item-row")
        count = 0
        for card in cards:
            try:
                name = card.find_element(By.CLASS_NAME, "item-title").text.strip()
                raw_address = card.find_element(By.CLASS_NAME, "address").text.strip()
                
                # تنفيذ التقسيم
                gov, dist, addr_details = process_address(raw_address)
                
                try:
                    # سحب التليفون لو متاح مباشرة
                    phone = card.find_element(By.CLASS_NAME, "phone-number").text.strip()
                except:
                    phone = "N/A"
                
                writer.writerow([name, gov, dist, addr_details, phone])
                count += 1
            except: continue
        return count
    except: return 0

# 3. اللوب الأساسي
base_url = "https://www.yellowpages.com.eg/ar/category/مطاعم"

try:
    for page in range(1, 51): # هيبدأ من صفحة 1 لـ 50
        target_url = f"{base_url}/p{page}"
        print(f"Scraping Page {page}...")
        
        driver.get(target_url)
        
        # وقت انتظار عشوائي طويل نسبياً لتجنب كشف البوت
        time.sleep(random.uniform(8, 12)) 
        
        # محاكاة حركة المستخدم (سكرول هادئ)
        for _ in range(3):
            driver.execute_script(f"window.scrollBy(0, 400);")
            time.sleep(random.uniform(1, 2))
        
        saved_count = scrape_data()
        print(f"Successfully saved {saved_count} items from page {page}.")
        
        if saved_count == 0:
            print("No data found or Blocked. Check the browser window!")
            break
            
        # استراحة قبل الصفحة التالية
        time.sleep(random.uniform(5, 7))

finally:
    csv_file.close()
    driver.quit()
    print("Mission Accomplished! Data is in 'yellowpages_final_separated.csv'")
```
